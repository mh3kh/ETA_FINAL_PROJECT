```python
import pandas as pd
```

```python
LIB = pd.read_csv("wikipedia-DOC_v2.csv").set_index('doc_id')
```

```python
LIB
```

```python
#check for any null values in doc_summary
LIB[LIB['doc_summary'].isnull()]
```

```python
#remove "Comparison of CDMI server implementations". The doc_summary is null because it is just a table on the wiki page. 

LIB = LIB.dropna()
```

```python
# Removes HTML tags and entities, then remaining LaTeX stuff,
# then removes newlines and collapses remaining spaces.

LIB['doc_summary_clean'] = LIB.doc_summary.str.replace(r"<[^>]+>", "", regex=True)\
    .str.replace(r"\sNaN\s", " ", regex = True)\
    .str.replace(r"\snan\s", " ", regex = True)\
    .str.replace(r"&\w+;", " ", regex=True)\
    .str.replace(r"\{\\\w+.+\}", " ", regex=True)\
    .str.replace(r"\n+", " ", regex=True)\
    .str.replace(r"\s+", " ", regex=True)\
    
```

```python
LIB['doc_content_clean'] = LIB.doc_content.str.replace(r"<[^>]+>", "", regex=True)\
    .str.replace(r"\sNaN\s", " ", regex = True)\
    .str.replace(r"\snan\s", " ", regex = True)\
    .str.replace(r"&\w+;", " ", regex=True)\
    .str.replace(r"\{\\\w+.+\}", " ", regex=True)\
    .str.replace(r"\n+", " ", regex=True)\
    .str.replace(r"\s+", " ", regex=True)
```

```python
LIB.head()
```

```python
#remove duplicate "abstract data types"
LIB["categories"] = LIB["categories"].replace('Abstract data types|Abstract data types', 'Abstract data types')
```

```python
LIB
```

# Save

```python
LIB.to_csv("wikipedia-DOC_v3.csv")
```
